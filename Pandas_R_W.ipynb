# Reading and writing  HTML Files

```python
frame = pd.DataFrame(np.arange(4).reshape(2,2))
frame
```

```python
print(frame.to_html())
```

```python
frame = pd.DataFrame(np.random.rand(4,4),
                     index =['white','black','red','blue'],
                     columns = ['up','down','right','left'])
frame
```

```python
# Web page
s = ['<HTML>']
s.append('<HEAD><TITLE>My DataFrame</TITLE></HEAD>')
s.append('<BODY>')
s.append(frame.to_html())
s.append('</BODY></HTML>')

html = "".join(s)
```

```python
html_file = open('myFrame.html','w')
```

```python
html_file.write(html)
html_file.close()
```

```python
# normally we need to convert the json file into tabular form. This process is called normalization
# function: json_nomalize()---- convert a json dict into a list.

from pandas.io.json import json_normalize
```

```python
#when the data is not tabular, we need to import json library, like this
import json
file = open('books.json','r')
text = file.read()
text = json.loads(text)
text
```

```python
## now, we are ready to apply json_normalize() function.
json_normalize(text, 'books')
```

```python
json_normalize(text,'books',['writer','nationality']) # writer and nationality aren't at the same level of books inside text.
```
